```python
import pandas as pd

# Load the dataset
file_path = "Raw-Vehicle-Crashes-NYC.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# List of columns to check
columns_to_check = ["CONTRIBUTING FACTOR VEHICLE 3", "CONTRIBUTING FACTOR VEHICLE 4", "CONTRIBUTING FACTOR VEHICLE 5", 
                    "VEHICLE TYPE CODE 3", "VEHICLE TYPE CODE 4", "VEHICLE TYPE CODE 5"]

# Remove rows where any of the specified columns have a value
df = df[df[columns_to_check].isna().all(axis=1)]

# Display the first few rows of the cleaned dataset
df.head()
```

```python
# Count the number of rows
num_rows = len(df)

print(f"The number of rows in the DataFrame is: {num_rows}")
```

```python
# List of columns to remove
columns_to_remove = ["ON STREET NAME", "CROSS STREET NAME", "OFF STREET NAME", 
                     "NUMBER OF PEDESTRIANS INJURED", "NUMBER OF PEDESTRIANS KILLED", 
                     "NUMBER OF CYCLIST INJURED", "NUMBER OF CYCLIST KILLED", 
                     "NUMBER OF MOTORIST INJURED", "NUMBER OF MOTORIST KILLED", 
                     "CONTRIBUTING FACTOR VEHICLE 3", "CONTRIBUTING FACTOR VEHICLE 4", 
                     "CONTRIBUTING FACTOR VEHICLE 5", "COLLISION_ID", 
                     "VEHICLE TYPE CODE 3", "VEHICLE TYPE CODE 4", "VEHICLE TYPE CODE 5"]

# Remove the columns
df = df.drop(columns=columns_to_remove)

# Display the first few rows of the cleaned dataset
df.head()
```

```python
# Remove rows with any missing values
df_clean = df.dropna()

# Print the number of remaining rows
print(f"The number of rows after removing rows with missing values: {len(df_clean)}")
```

```python
df_clean.head()
```

```python
# Convert "CRASH DATE" to datetime format
df_clean.loc[:, 'CRASH DATE'] = pd.to_datetime(df_clean['CRASH DATE'])

# Combine "CRASH DATE" and "CRASH TIME" into a single datetime column
df_clean.loc[:, 'CRASH DATETIME'] = pd.to_datetime(df_clean['CRASH DATE'].dt.date.astype(str) + ' ' + df_clean['CRASH TIME'])

# Sort the DataFrame by the new datetime column
df_clean = df_clean.sort_values(by='CRASH DATETIME')

# Display the first few rows of the sorted DataFrame
df_clean.head()
```

```python
# Remove "CRASH DATE" and "CRASH TIME" columns
df_clean = df_clean.drop(columns=['CRASH DATE', 'CRASH TIME'])

# Display the first few rows of the cleaned DataFrame
df_clean.head()
```

```python
# Move "CRASH DATETIME" column to the first position
col_name = "CRASH DATETIME"
first_col = df_clean.pop(col_name)
df_clean.insert(0, col_name, first_col)

# Display the first few rows of the DataFrame
df_clean.head()
```

```python
# Remove "LOCATION" column
df_clean = df_clean.drop(columns=['LOCATION'])

# Display the first few rows of the cleaned DataFrame
df_clean.head()
```

```python
# Count the number of rows
num_rows = len(df_clean)

print(f"The number of rows in the DataFrame is: {num_rows}")
```

```python
# Define the file path
file_path = "Data/Cleaned-Vehicle-Crashes-NYC.csv"

# Export the DataFrame to a CSV file
df_clean.to_csv(file_path, index=False)
```
